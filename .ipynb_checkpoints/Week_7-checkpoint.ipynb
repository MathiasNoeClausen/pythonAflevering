```python
import bs4

import requests


r = requests.get('https://www.dba.dk/biler', headers={'User-Agent': 'Mozilla/5.0'})
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')

print(soup.prettify()[:1500])
```

```python
import requests
from bs4 import BeautifulSoup
req = requests.get('https://www.dba.dk/biler', headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(req.text, 'lxml')
print(soup.prettify()[:1000])
print('------------------------')
for link in soup.find_all('a'):
    if not link.get('href').startswith('/'):
        continue
    print(link.get('href'))
```
