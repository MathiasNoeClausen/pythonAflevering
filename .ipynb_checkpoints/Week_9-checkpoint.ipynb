```python
# Download the data
import numpy as np
import pandas as pd
url = 'https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Feature engineering
from sklearn import preprocessing
mask = (df['Identification.Make'] == 'Honda') & (df['Fuel Information.Fuel Type'] == 'Gasoline')
df_mask = df[mask]
df_mask
# Tror umiddelbart dette er rigtigt, men er ikke sikker. Finder de nederste 90% af af honda biler med br√¶ndstof typen gasoline ud fra deres horsepower
quantiles = df_mask['Engine Information.Engine Statistics.Horsepower'] < df_mask['Engine Information.Engine Statistics.Horsepower'].quantile(0.9)
df1 = df_mask[quantiles]

scaler = preprocessing.MinMaxScaler()
#names = quantiles.columns
d = scaler.fit_transform(df1)
scaled_df = pd.DataFrame(d)
scaled_df.head()
```
